# Gradient Descent for Linear Regression ‚Äî Theory

## üìå Problem Statement

We are given a dataset containing a relationship between **Hours Studied** and **Exam Score**. The goal is to fit a **straight line** of the form:

$$
\hat{y} = mx + b
$$

Where:
- $x$: Number of hours studied (independent variable)  
- $\hat{y}$: Predicted exam score (dependent variable)  
- $m$: Slope of the regression line (how much the score changes with each additional hour studied)  
- $b$: Intercept (predicted score when $x = 0$)

---

## üß† Loss Function: Mean Squared Error (MSE)

To evaluate how well the predicted line fits the actual data, we use the **Mean Squared Error (MSE)**, defined as:

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 = \frac{1}{n} \sum_{i=1}^{n} \left( y_i - (mx_i + b) \right)^2
$$

This measures the average squared difference between actual values \( y \) and predicted values \( \hat{y} \).

---

## üîÑ Optimization: Gradient Descent

To minimize the MSE, we use **gradient descent**, an iterative optimization technique that updates model parameters to reduce error.

The gradients (partial derivatives) of MSE with respect to \( m \) and \( b \) are:

$$
\frac{\partial}{\partial m} = \frac{-2}{n} \sum x_i(y_i - \hat{y}_i)
$$

$$
\frac{\partial}{\partial b} = \frac{-2}{n} \sum (y_i - \hat{y}_i)
$$

These gradients indicate the direction in which \( m \) and \( b \) should be adjusted to reduce the error.

---

## üîÅ Parameter Update Rules

In each iteration of gradient descent, we update the slope and intercept using:

$$
m = m - \alpha \cdot \frac{\partial \text{MSE}}{\partial m}
$$

$$
b = b - \alpha \cdot \frac{\partial \text{MSE}}{\partial b}
$$

Where:
- $\alpha$ is the **learning rate**, a small constant that controls how large each update step is.  
- This process repeats over multiple **epochs** (iterations) until convergence.


---

## ‚úÖ Summary

- Linear regression tries to fit a line to the data using the equation \( y = mx + b \).
- The goal is to minimize the MSE using gradient descent.
- The gradients tell us how to adjust \( m \) and \( b \).
- Repeated updates gradually lead to a line that best fits the data.

---


```python
import pandas as pd

data = pd.read_csv('sample_linear_regression_dataset.csv')

def mean_squared_error(m, b, points):
    total_error = 0

    for i in range(len(points)):
        x = points.iloc[i].Hours_Studied
        y = points.ilove[i].Exam_Score
        total_error += (y - (m * x + b)) ** 2 

    total_error / float(len(points))

def gradient_descent(m_now, b_now, points, alpha=0.0001):
    m_gradient = 0
    b_gradient = 0

    for i in range(n := len(points)):
        x = points.iloc[i].Hours_Studied
        y = points.iloc[i].Exam_Score

        m_gradient += -(2/n) * x * (y - (m_now * x + b_now))
        b_gradient += -(2/n) * (y - (m_now * x + b_now))

        m = m_now - m_gradient * alpha
        b = b_now - b_gradient * alpha

    return m, b
        
```

```python
import matplotlib.pyplot as plt
m = 0
b = 0
epochs = 500

# Training loop
for i in range(epochs):
    if i % 50 == 0:
        print(f"Epoch: {i}")
    m, b = gradient_descent(m, b, data)

# Output results
print(f"Final Slope (m): {m:.4f}, Intercept (b): {b:.4f}")

# Plotting
plt.scatter(data.Hours_Studied, data.Exam_Score, color="black")
x_vals = list(range(0, 15))
y_vals = [m * x + b for x in x_vals]
plt.plot(x_vals, y_vals, color="red")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("Linear Regression via Gradient Descent")
plt.grid(True)
plt.show()
```
